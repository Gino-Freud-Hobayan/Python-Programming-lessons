# 6. Sets


```python
## Somewhat similar to Lists/Tuples, but THEY DON'T HAVE DUPLICATE ELEMENTS
## Values in a set cannot be accessed using an index, because it doesn't have an index = because it's not ordered.
```

```python
daily_pints = {1,2,3}
```

```python
type(daily_pints)
```

```python
print(daily_pints)
```

```python
## Only the Unique values will appear if you run this.


daily_pints_log = {1,2,31,2,3,4,1,2,5,6,3,2}


print(daily_pints_log)
```

```python
## Uses of Set = COMPARING 2 SEPARATE SETS.
## Comparing the unique values in 1 set to the unique values in another set.
```

```python

## EX: Comparing my daily_pints_log with my wifes_daily_pints_log

wifes_daily_pints_log = {1,3,5,7,3,24,5,7,3,2,0}
```

```python
## The result is = the unique values between these 2 datasets.

print(daily_pints_log | wifes_daily_pints_log)
```

```python
## The result is = Which one shows up in both datasets.

print(daily_pints_log & wifes_daily_pints_log)
```

```python
## The result is = Which one DOESN'T MATCH in both datasets.

print(daily_pints_log - wifes_daily_pints_log)
```

```python
## The result is = Which one DOESN'T MATCH in both datasets.

print(wifes_daily_pints_log - daily_pints_log )
```

```python
## The result is = it will show if the value is in one or the other, but not in both.
# These values are completely unique to each of their own datasets

print(wifes_daily_pints_log ^ daily_pints_log )
```

# 7. Dictionary

```python
## Most different out of the previous data types


## Key/value Pair 
## In a dictionary, it's not like a list, where you have 1,2,3,4 (comma,value,comma,value)


## we also use {} like set. The difference is, Sets are just comma separated values while Dictionary has Key/value pairs.
```

```python
## We can also add a list inside this dictionary, check out: favorite ice creams
## Key/Value Pair

dict_cream = {'name': 'Gino Hobayan', 'weekly intake': 5, 'favorite ice creams':['MCC', 'Chocolate'] }
```

```python
type(dict_cream)
```

```python
print(dict_cream)
```

```python
## When we execute this code, we will see all of the values in this dictionary.

dict_cream.values()
```

```python
## When we execute this code, we will see all of 

dict_cream.keys()
```

```python
## When we execute this code, we will see all of the Key/Value pairs

dict_cream.items()
```

```python
## The main difference between a List and a Dictionary is how you call the index



dict_cream[0]
```

```python
## You have to specify the index.

dict_cream['name']
```

```python
## You can also update/CHANGE some information/values on a Dictionary


dict_cream['name'] = 'Gino Gino Gino'

print(dict_cream)
```

```python
## UPDATE
## What we're going to do now is to update this entire thing.

dict_cream.update({'name': 'Gino Gino Gino', 'weekly intake': 5, 'weight': 300})


print(dict_cream)
```

## It did not delete the previous Key/Value pair. Instead, it just added to it.

```python
## DELETE
## What we're going to do now is to DELETE

del dict_cream['weight']


print(dict_cream)
```
