# If - Elif - Else Statements



# IF Statement

### If this IF Statement is TRUE = We are going to run a body of code (body of if)
### If this IF Statement is FALSE = We are going to the ELIF condition (elif condition)


# ELIF Statement 
## = If the First IF Statement doesn't work, Let's try this IF Statement instead.

.
.

### If this ELIF Statement is TRUE = We are going to run a body of code (body of elif)
### If this ELIF Statement is FALSE = We are going to the body of else (body of else)




# ELSE Statement

## = If the IF and ELIF statement doesn't work, then run this body of code



# Examples:


```python
## If the statement of 25>10 is TRUE, then this code will print the string: 'it worked'


if 25 > 10:
    print('It worked')
```

```python
## If the statement of 25<10 is FALSE, then this code will not print.
## If we use an ELSE statement = 


if 25 < 10:
    print('it worked')
else:
    print('It did not work')
```

## What if we have a lot of different conditions we want to try?
## ELIF statement

```python
## 1.) It's going to check if 25 is less than 10 (FALSE)
## 2.) Is 25 less than 30? (TRUE) 
## And if 2.) is TRUE, it will print 'elif worked'



if 25 < 10:
    print('it worked')
elif 25 < 30:
    print('elif worked')
else:
    print('It did not work')
```

## Multiple ELIF statements
## Once elif hits something TRUE (elif 25 < 40 = 'elif 3 worked') none of the other elif or else statement is going to print.

```python
if 25 < 10:
    print('it worked')
    
elif 25 < 20:
    print('elif worked')
elif 25 < 21:
    print('elif 2 worked')
elif 25 < 40:
    print('elif 3 worked')     ## It ends here in this line.
elif 25 < 50:
    print('elif 4 worked')
    
else:
    print('It did not work')
```

## We can also use a LOGCIAL OPERATOR (and, or)


```python
if (25 < 10) or (1<3):
    print('it worked')
    
elif 25 < 20:
    print('elif worked')
elif 25 < 21:
    print('elif 2 worked')
elif 25 < 40:
    print('elif 3 worked')
elif 25 < 50:
    print('elif 4 worked')
    
else:
    print('It did not work')


    
## The first IF statement worked since we used the OR logical operator and 1<3 is TRUE.
```

# Nested IF/else statement

## Multiple IF Statements = It's still the same body of code, so it goes down to the 2nd IF statement.


#### You can do this on and on and on, it can basically go on forever
#### and you can create a really in-depth logic and that actually happens a lot when you start writing more advanced code 


```python
if (25 < 10) or (1<3):
    print('The 1st IF statement worked!')
    if 10 > 5:
        print('This nested IF statement worked!')
    
    
elif 25 < 20:
    print('elif worked')
elif 25 < 21:
    print('elif 2 worked')
elif 25 < 40:
    print('elif 3 worked')
elif 25 < 50:
    print('elif 4 worked')
    
    
else:
    print('It did not work')
```
