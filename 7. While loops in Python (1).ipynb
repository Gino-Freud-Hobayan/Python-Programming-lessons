# While Loops

### Used to iterate over a block of code as long as the Test condition is True



### The Main Difference between "For Loops" and "While Loops" is that 

### 1.) For loops will loop over the entire sequence REGARDLESS OF THE CONDITION

### 2.) While Loops will only iterate over that sequence AS LONG AS A SPECIFIC CONDITION IS MET.

```python
number = 0

while number < 5:                 ## In English = While the number is less than 5, print the number
    print(number)
    number = number + 1           ## We are creating a Counter, so we can count the no. of reps
    
```

```python
## In English = While the number is less than 5, print the number
## Once this number is no longer less than 5, it will BREAK OUT of the WHILE loop and it will end.
## It starts at 0 and the loop ends at 4.

## It never goes back to the 'number = 0'
```

# BREAK statement

### = We can stop the While loop even if the Test Condition is TRUE

```python
number = 0

while number < 5:
    print(number)
    if number == 3:                  ## If the number is equal to 3, we are going to Break out of the loop.
        break
    number = number + 1
```

# ELSE Statement

```python
number = 0

while number < 5:
    print(number)
    if number == 6:                          
        break
    number = number + 1            ## Once this condition is no longer met, it will go to the ELSE Statement
    
else:
    print('No longer < 5')         ## The ELSE Statement is only going to trigger once the while loop is no longer TRUE.
```

```python
## In this case, it never reaches the ELSE Statement because of the BREAK statement == 3

number = 0

while number < 5:
    print(number)
    if number == 3:                          
        break
    number = number + 1            ## Once this condition is no longer met, it will go to the ELSE Statement
    
else:
    print('No longer < 5')         ## The ELSE Statement is only going to trigger once the while loop is no longer TRUE.
                                   ## We changed the IF number = 3, so the While Loop is no longer true
                                   ## and it never reached the ELSE statement.
```

# Continue Statement

```python
## Let's change the sequence


number = 0

while number < 5:
    number = number + 1 
    if number == 3:                          
        continue                 ## English = If the number is 3, then Continue and print the number
    print(number)
          
    
else:
    print('No longer < 5')
```

```python
## INFINITE LOOP

number = 0

while number < 5:
    print(number)
    if number == 3:                          
        continue                      ## This will create an INFINITE LOOP of 3 
    number = number + 1               ## and disregard our counter for reps below it.
    
else:
    print('No longer < 5')
```
