# For loops in a LIST

```python
## Here's the List we will be working with in this lesson.

integers = [1,2,3,4,5]
```

```python
## During the 1st loop = 1 will be printed (number), 
## the next iteration/loop = 2 will be printed (number)
## the next iteration/loop = 3 will be printed (number)
```

```python
## Syntax


for number in integers:
    print(number)
```

### We are not really using the numbers inside the list, 
### WE'RE JUST USING IT LIKE A COUNTER for how many repetitions. 1,2,3,4,5,.....

```python
integers = [1,2,3,4,5]

for number in integers:
    print('yep!')
```

#### The output is Jelly + Jelly. 
#### On the 1st loop, Jelly = 1 (1+1=2)
#### On the 2nd loop, Jelly = 2 (2+2=4)

```python
integers = [1,2,3,4,5]

for Jelly in integers:
    print(Jelly + Jelly)
```

# For loops in a DICTIONARY

```python
ice_cream_dict = {'name': 'Gino Hobayan', 'weekly intake': 5, 'favorite ice creams': ['MCC', 'Chocolate']}

## We have a list inside of this dictionary = : ['MCC', 'Chocolate']
```

```python
## .values() = used to specify WITHIN the dictionary what we want to print out 
## (Are we pulling the item?, the value?, etc.)



for cream in ice_cream_dict.values():
    print(cream)
```

```python
## Pulling KEY 


for key in ice_cream_dict.items():
    print(key)
```

## Pulling both KEY and VALUE.


```python
ice_cream_dict = {'name': 'Gino Hobayan', 'weekly intake': 5, 'favorite ice creams': ['MCC', 'Chocolate']}



for key, value in ice_cream_dict.items():
    print(key, '->', value)
```

# Nested 'For Loops'

###  For loop in a For loop

```python
## Let's create 2 separate lists

flavors = ['Vanilla', 'Chocolate', 'Cookie Dough']
toppings = ['Hot Fudge', 'Oreos', 'Marshmallows']
```

### What's going to happen = 
### We are going to take the one in flavors (Vanilla) and It's going to loop to the entire roster of the toppings list.


```python

for one in flavors:
    for two in toppings:
        print(one, 'topped with', two)
```

```python
### The Output is 9 different results.
```
