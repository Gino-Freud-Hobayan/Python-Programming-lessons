# 2. Logical Operators

```python
## We have a TRUE statement and a FALSE statement
## We get a FALSE since it needs BOTH STATEMENTS TO BE TRUE.


(10>50) and (50>10)
```

```python
## Both TRUE statements

(70>50) and (50>10)
```

```python
## We have a TRUE statement and a FALSE statement
## Even if one of them is a TRUE statement, the Output will be TRUE.

(10>50) or (50>10)
```

## Vanilla is actually greater than Chocolate since V is a Higher number in the Alphabetical order

```python
('Vanilla' > 'Chocolate') or (50>10)

('Vanilla' > 'Chocolate') and (50>10)
```

```python
not(50>10)
```

# 3. Membership Operators

```python
ice_cream = 'I love Chocolate ice cream'

'love' in ice_cream
```

## Basically just a Find or Ctrl + F on Python.
## We are searching for a specific value.

```python
## We can also make a List
## All we're doing is searching if 2 is in the list, and if 6 is not in the list.

scoops = [1,2,3,4,5]



2 in scoops

6 not in scoops
```

```python
## Since wanted_scoops has been assigned the number 8.
## So the statement wanted_scoops not in scoops is TRUE

scoops = [1,2,3,4,5]
wanted_scoops = 8

wanted_scoops not in scoops
```

```python
## Since wanted_scoops has been assigned the number 8.
## So the statement wanted_scoops in scoops is FALSE

scoops = [1,2,3,4,5]
wanted_scoops = 8

wanted_scoops in scoops
```
