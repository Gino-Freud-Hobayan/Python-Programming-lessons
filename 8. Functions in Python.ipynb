# Argument

### Information can be passed into functions as arguments.

### Arguments are specified after the function name, inside the parentheses. You can add as many arguments as you want, just separate them with a comma.

```python

def number_squared(number):
    print(number**2)                     ## It's going to take the 'number' and put it to the power of 2 / Square it.
    

    
number_squared(5)
```

```python
number_squared(6)
```

```python
number_squared(12)
```

# Multiple Arguments

```python
## Now we can specify our number, and the power we want to raise is to, not just the default of 2.


def number_squared_custom(number,power):
    print(number**power) 
    


    
## When we call this function, we can specify the number and the power

## 5 to the power of 3
number_squared_custom(5,3)
```

```python
## 3 to the power of 3
number_squared_custom(3,3)
```

```python
## 2 to the power of 3
number_squared_custom(2,3)
```

# Arbitrary Arguments, *args

### If you do not know how many arguments that will be passed into your function, add a * before the parameter name in the function definition.

#### This way the function will receive a tuple of arguments, and can access the items accordingly:





```python
def number_args(*args):                   ## We're putting *args if we still dont know how many 
    
```

```python
def number_args(*number):                 ## We're putting *number
    print(number[0]*number[1])


## When we call out the function 'number_args', we will need to SPECIFY the number [0] and number [1]
```

```python
## print(number[0]*number[1]) = It will multiply the number 0 and number 1 in the list, in this case = 5 and 6

number_args(5,6,1,2,8)
```

```python
## print(number[0]*number[1]) = It will multiply the number 0 and number 1 in the list, in this case = 6 and 6

number_args(6,6,1,2,8)
```

```python
args_tuple = (5,6,1,2,8)

def number_args(*number):                 
    print(number[0]*number[1])
```

```python
number_args(args_tuple)                        ## This Failed because it doesnt have an *args_tuple
```

```python
number_args(*args_tuple)                        ## CORRECT. We need to use the * before the parameter.
```

# Keyword Arguments
### You can also send arguments with the key = value syntax.

### This way the order of the arguments does not matter.

```python
def number_squared_custom(number,power):
    print(number**power) 
```

```python
## If you want more control, you can use Keyword Arguments, we just switched the power and number
```

```python
## 3 to the power of 5 = 243

number_squared_custom(power = 5, number = 3)  
```

# Arbitrary Keyword Arguments, **kwargs
### If you do not know how many keyword arguments that will be passed into your function, add two asterisk: ** before the parameter name in the function definition.

This way the function will receive a dictionary of arguments, and can access the items accordingly:

```python
## Arbitrary = We don't know how many keyword arguments we want to pass into our function.
```

```python
## You're gonna need to go back and forth between the Call function and the print.


def number_kwarg(**number):
    print('My number is: ' + number['integer'])     ## 2.)  then go back to print to finish it
```

```python
number_kwarg(integer='2309')                        ## 1.) Create the value of integer first (integer='2309')
```

```python
## Using 2 Arbitrary Keyword Arguments



def number_kwarg(**number):
    print('My number is: ' + number['integer'] + ' My other number: ', number['integer2'])
```

```python
number_kwarg(integer='2309', integer2 = '349')
```
